```python
train = pd.read_csv('../input/msbd5001-spring-2022/train.csv')
train
```

```python
clf = RandomForestClassifier(max_depth=10,random_state=0)
```

```python
train_x = train[['MO HLADR+ MFI (cells/ul)', 'Neu CD64+MFI (cells/ul)',
       'CD3+T (cells/ul)', 'CD8+T (cells/ul)', 'CD4+T (cells/ul)',
       'NK (cells/ul)', 'CD19+ (cells/ul)', 'CD45+ (cells/ul)', 'Age',
       'Sex 0M1F', 'Mono CD64+MFI (cells/ul)']]
train_x
```

```python
im = SimpleImputer(missing_values = np.nan,strategy = 'mean')
train_x = im.fit_transform(train_x)
```

```python
train_y = train['label']
```

```python
clf.fit(train_x,train_y)
```

```python
test = pd.read_csv('../input/msbd5001-spring-2022/test.csv')
```

```python
res = clf.predict(test_x)
```
